```python
b = pd.read_csv('bookings.csv.bz2', 
                sep='^', 
                nrows=1e5, 
                usecols=['brd_time           ','brd_port','off_port'],
                parse_dates = ['brd_time           '])
b['dateNoTimeDate'] = b['brd_time           '].apply(lambda d: d.date())
b.head()
```

```python
s = pd.read_csv('searches.csv.bz2', 
                sep='^', 
                nrows=1e5, 
                usecols=['Origin','Destination','Seg1Date'],
                parse_dates=['Seg1Date'])
s['dateNoTimeDate'] = s['Seg1Date'].apply(lambda d: d.date())
s.head(5)
```

```python
m = pd.merge(s.dropna(),
             b.dropna(),
             how='left',
             right_on=['brd_port','off_port','dateNoTimeDate'],
             left_on=['Origin','Destination','dateNoTimeDate']
            )
```

```python
m['match'] = m.brd_port.apply(lambda b: not pd.isnull(b))
m.head(10)
```

```python
# mal porque hay espaciuos en los nombres de aeropuertos
m.groupby('match').count()
```
