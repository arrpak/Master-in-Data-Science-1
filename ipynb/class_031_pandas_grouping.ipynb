### Merging Dataframes

```python
df1 = DataFrame(np.random.randint(0,10,3*5).reshape(3,5))
df1.columns = ['c1','c2','c3','c4','c5']
df1['key1'] = ['a','a','b']
df1
```

```python
df2 = DataFrame(np.random.randint(0,10,3*5).reshape(3,5))
df2.columns = ['c1','c2','c3','c4','c5']
df2['key1'] = ['b','a','b']
df2
```

```python
# pandas tries join using the first matched column
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on="key1")
```

```python
pd.merge(df1,df2,on="key1",how="inner")
```

```python
pd.merge(df1,df2,on="key1",how="outer")
```

```python
df2["key2"]=df2["key1"]
df2
```

```python
pd.merge(df1,df2,left_on="key1",right_on="key2",suffixes=("izq","der"))
```

```python
df2 = df2.append({'c1':2,'c2':3,'c3':4,'c4':5,'c5':6,'key1':'d','key2':'c'},ignore_index=True)
df2
```

```python
pd.merge(df1,df2,left_on="key1",right_on="key2",how="inner")
```
